```python
###############################
# init a and b - choose random values - N(0,1)
# define epochs=20, L=0.05
# iterate over dataset to cacluate new a,b (by formula)
# plot the loss (error / cost) over epochs
# what will bhappen if you'll add 1000 to the X's. why?
```

```python
np.random.seed(47)
N = 150
X = np.random.rand(N)
```

```python
a = 1.5
b = 2.5
```

```python
y_hat = a*X +b
```

```python
np.random.seed(47)
a = np.random.rand(1)
b = np.random.rand(1)
print(a,b)

L=0.05
epochs=1000

l = []
```

```python
for epoch in range(epochs):
    y_hat = a*X +b
    e =  y- y_hat
    loss = (e**2).mean()
    l.append(loss)
    dLda = -2*((e*X).mean())
    dLdb = -2*(e.mean())
    a = a- L*dLda
    b = b- L*dLdb
print(a,b)
```

```python
plt.plot(np.log10(l[:100]))
```
