```python
import statsmodels.api as st
import numpy as np
import pandas as pd
```

```python
# !pip install --upgrade statsmodels
```

```python
df = pd.read_csv('pupils.csv')
```

```python
df.head()
```

```python
x = df.drop(['Avg Grades','Name','Country'],axis=1)
y = df['Avg Grades']
```

```python
x.head()
```

```python
x1 = st.add_constant(x)
x1.head(2)
```

```python
# res = st.OLS(y,x1).fit()
```

```python
model = st.OLS(y,x1)
```

```python
res = model.fit()
```

```python
res.params
```

```python
res.summary()
```

```python
res.predict([1,7,120,30,30000,5,6])
```
