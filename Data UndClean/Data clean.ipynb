```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

# dealing with missing values

```python
df = pd.read_csv('dataclean.csv')
df.head()
```

```python
df.B.hist(bins=15)
```

```python
df.info()
```

```python
sb.distplot(df.B[df.B.notnull()],bins=10)
```

```python
sb.distplot(df['B'],bins=10)
```

```python
s1 = pd.Series(np.random.normal(50,10,100))
df['B'].fillna(s1,inplace=True)
```

```python
sb.distplot(s1,bins=20)
```
