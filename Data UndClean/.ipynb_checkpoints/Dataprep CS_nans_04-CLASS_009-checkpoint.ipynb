```python
df.Height.mean()
```

```python
df.Height.fillna(df.Height.mean(),inplace=True)
```

```python
#deciding based on business knowledge
```

```python
df.loc[df.Age > 12,'Age'] = 12
```

```python
df.isna().sum()
```

```python
# ALTERNATIVE FOR FILL NULL - by some formulas
df.loc[df.Height.isnull(),'Height'] = df.Age*12
```

```python
df.drop(df.loc[df.Age <= 8].index,inplace=True)
```

```python
df.describe().T
```

```python
df.drop(['type'],axis=1,inplace=True)
```

```python
df['namelen']=df.Name.apply(len)
```

```python
#alternative
df['namelen2']=df.Name.str.len()
```

```python
# BINNING
```

```python
def myop(inc):
    if inc < 10000:
        return 1
    else:
        if inc < 30000:
            return 2
        else:
            return 3
```

```python
df['new_inc'] = df.income.apply(myop)
```

```python
df.head()
```

```python
#####################
# lambda the above (apply with lambda)
# verify same values
```

```python
df['new_inc2'] = df.income.apply(lambda x: 1 if x < 10000 else (2 if x < 20000 else 3))
```

```python
df['new_inc2'] = df.income.apply(lambda x: 1 if x < 10000 else 2 if x < 30000 else 3)
```
