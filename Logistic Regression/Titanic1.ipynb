```python
cmat = skmet.confusion_matrix(y_test,predictions)
```

# TN       FP
# FN       TP

```python
cmat
```

precision - Actual yes / Total perdicted yes - TP / (FP + TP)

```python
pre = cmat[1,1]/(cmat[0,1]+cmat[1,1])
pre
```

recall - TP / (FN + TP)

```python
recall = cmat[1,1]/(cmat[1,0]+ cmat[1,1])
recall
```

```python
f1_score = 2*((pre*recall)/(pre + recall))
f1_score
```
